```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data
```

```python
x=data.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import OPTICS
opticsClus= OPTICS(min_samples=10, xi=0.05, min_cluster_size=0.05)
cluster=opticsClus.fit_predict(x)
```

```python
cluster
```

```python
supervised=pd.DataFrame(data)
supervised['cluster_group']=cluster
```

```python
supervised
```

```python
supervised.to_csv('OPTICS.csv',index=False)
```
